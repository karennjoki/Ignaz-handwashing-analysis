```python
yearly = pd.read_csv(r'C:\Users\KAREN\Documents\Ignaz Handwashing Dataset\yearly_deaths_by_clinic.csv')
print(yearly)
```

```python
yearly['proportion_deaths'] = yearly['deaths'] / yearly['births']
clinic_1 = yearly[yearly['clinic'] == 'clinic 1']
clinic_2 = yearly[yearly['clinic'] == 'clinic 2']
print(clinic_1)
print(clinic_2)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(clinic_1.year, clinic_1.proportion_deaths, label = 'clinic1')
plt.plot(clinic_2.year, clinic_2.proportion_deaths, label = 'clinic2')
plt.xlabel('Year')
plt.ylabel('Proportion Deaths')
plt.legend()
```

```python
monthly = pd.read_csv(r'C:\Users\KAREN\Documents\Ignaz Handwashing Dataset\monthly_deaths.csv',  parse_dates =['date'])
monthly['proportion_deaths'] = monthly['deaths'] / monthly['births'] 
print(monthly)
```

```python
plt.plot(monthly.date, monthly.proportion_deaths)
plt.xlabel('Date')
plt.ylabel('Proportion Deaths')
plt.show
```

```python
handwashing_start = pd.to_datetime('1847-06-01')
before_washing = monthly[monthly['date'] < '1847-06-01' ]
after_washing = monthly[monthly['date'] > '1847-06-01' ]
print(before_washing)
print(after_washing)
```

```python
plt.plot(before_washing.date, before_washing.proportion_deaths, label = 'Before Washing')
plt.plot(after_washing.date, after_washing.proportion_deaths, label = 'After Washing')
plt.xlabel('date')
plt.xlabel('proportion_deaths')
plt.legend()
```

```python
# Difference in mean monthly proportion of deaths due to handwashing
before_proportion = before_washing.proportion_deaths
after_proportion = after_washing.proportion_deaths
mean_diff = after_proportion.mean() - before_proportion.mean()
mean_diff
```

```python
# A bootstrap analysis of the reduction of deaths due to handwashing
boot_mean_diff = []
for i in range(3000):
    boot_before = before_proportion.sample(frac=1, replace=True)
    boot_after = after_proportion.sample(frac=1, replace=True)
    boot_mean_diff.append( boot_after.mean() - boot_before.mean() )

# Calculating a 95% confidence interval from boot_mean_diff 
confidence_interval = pd.Series(boot_mean_diff).quantile([0.025, 0.975])
confidence_interval
```
